```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cohort = pd.read_excel('cohort_start_year.xlsx')
```

```python
cohort.head()
```

```python
cohort_sample = cohort[cohort['first_year'] >= 2012]
```

```python
cohort_sample.shape
```

```python
sns.set_theme(style='darkgrid')
sns.lineplot(
    data=cohort_sample,
    x='period',
    y='pct_retained',
    hue='first_year',
    palette=sns.color_palette(n_colors=21)
)
plt.savefig('cohort.png')
```
